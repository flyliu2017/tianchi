```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
path="/root/PycharmProjects/tf/data"
```

```python
raw_data=pd.read_csv(path+"/шонч╗Г.csv")
```

```python
raw_data.keys()
```

```python
raw_data=raw_data.set_index('ID')
```

```python
raw_data.index
```

```python
column_counts=raw_data.count()
```

```python
data_valid=raw_data[column_counts[column_counts>500*0.7].keys()]
```

```python
data_valid.keys()
```

```python
missing_cols=raw_data.drop(data_valid.keys(),axis=1)
```

```python
missing_cols
```

```python
data_valid
```

```python
test_a=pd.read_excel(path+"/ц╡ЛшпХA.xlsx")
test_b=pd.read_excel(path+"/ц╡ЛшпХB.xlsx")
```

```python
test_a=test_a.set_index('ID')
```

```python
test_b=test_b.set_index('ID')
```

```python
combined_data=pd.concat([data_valid,test_a,test_b],join='inner')
combined_data
```

```python
tool_cols=combined_data.filter(regex='[oO]')
tool_cols
```

```python
tool_cols.dtypes
```

```python
combined_data[tool_cols.keys()]=combined_data[tool_cols.keys()].astype('str')
```

```python
combined_data[tool_cols.keys()].dtypes
```

```python
one_hoted_combined_data=pd.get_dummies(combined_data)
```

```python
one_hoted_combined_data
```

```python
cv=one_hoted_combined_data.std()/one_hoted_combined_data.mean()
cv=cv.abs()
```

```python
cv
```

```python
one_value_cols=one_hoted_combined_data[cv[cv<1e-8].keys()]
```

```python
one_value_cols
```

```python
delete_one_value=one_hoted_combined_data.drop(one_value_cols.keys(),axis=1)
delete_one_value
```

```python
imputer=preprocessing.Imputer(strategy='median')
```

```python
filled=imputer.fit_transform(delete_one_value)
```

```python
filled
```

```python
valid_combined_data=pd.DataFrame(filled,index=delete_one_value.index,columns=delete_one_value.keys())
```

```python
valid_combined_data
```

```python
train_data=valid_combined_data.iloc[0:500]
```

```python
train_data
```

```python
test_a_data=valid_combined_data.iloc[500:600]
test_a_data
```

```python
test_b_data=valid_combined_data.iloc[600:]
test_b_data
```

```python
train_data.to_csv(path+'/train_data.csv')
test_a_data.to_csv(path+'/test_a_data.csv')
test_b_data.to_csv(path+'/test_b_data.csv')
```
